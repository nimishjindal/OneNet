```python
import glob

base = "price_long_50/"

files = glob.glob(base+'*')
files
```

```python
import pandas as pd
```

```python
all_dataframes = []
target = "Close"
target_column_name = 'Target'

columns = ["Date", "High", "Low", "Open" ,"Volume", "Close"]
columns_w_target = columns + [target_column_name]

for file in files:
    df = pd.read_csv(file, parse_dates=['Date'], usecols=columns)
    features = df[1:].copy()
    target_df = df[target].shift(1).copy()
    features[target_column_name] = target_df
    print(file + " size: " + str(len(features)))
    
    all_dataframes.append(features.copy())

final_dataframe = pd.concat(all_dataframes, ignore_index=True)
```

```python
final_dataframe = final_dataframe[columns_w_target]
final_dataframe.head()
```

```python
final_dataframe.to_csv('transformed/all_companies_concat.csv', index=False)
```
